```python
# IMPORT LIBRARY
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import KFold, cross_validate
```

```python
# IMPORT DATASET
data = pd.read_csv('scores.csv', sep=';')
data.head()
```

```python
# MENENTUKAN DATA TRAINING SEBANYAK 2O BARIS PERTAMA
X_train = data.iloc[:20,1:6].values
y_train = data.iloc[:20,6].values
```

```python
# MENENTUKAN DATA TESTING SEBANYAK 20 BARIS TERAKHIR
X_test = data.iloc[20:,1:6].values
y_test = data.iloc[20:,6].values
```

```python
# DEKLARASI VARIABEL ESTIMATOR SEBAGAI PENAMPUNG SUB MODEL 3 ALGORITMA
estimators = []

# MEMBUAT SUB MODEL MENGGUNAKAN 3 ALGORITMA
# MODEL MENGGUNAKAN KNN
model1 = KNeighborsClassifier()
model1.fit(X_train, y_train)
pred1 = model1.predict(X_test)
estimators.append(('KNN', model1))

# MODEL MENGGUNAKAN NAIVE BAYES
model2 = GaussianNB()
model2.fit(X_train, y_train)
pred2 = model2.predict(X_test)
estimators.append(('Naive Bayes', model2))

# MODEL MENGGUNAKAN SVM
model3 = SVC(gamma=1)
model3.fit(X_train, y_train)
pred3 = model3.predict(X_test)
estimators.append(('SVM', model3))
```

```python
# MODEL ENSEMBLE DENGAN METODE VOTING
ensemble = VotingClassifier(estimators)
ensemble.fit(X_train, y_train)
pred4 = ensemble.predict(X_test)

# MENAMPILKAN PREDIKSI KEDALAM DATA FRAME
cols = {
    'Y True' : y_test,
    'Pred KNN' : pred1,
    'Pred Naive Bayes' : pred2,
    'Pred SVM' : pred3,
    'Voting' : pred4

}
print("Prediksi pada setiap Algoritma")
pd.DataFrame(cols)
```

```python
# MENGHITUNG PERFORMANCE PREDIKSI MODEL ENSEMBEL
kfold = KFold(n_splits=5)
scores = ['accuracy', 'precision', 'recall']
results = cross_validate(ensemble, X_test, y_test, cv=kfold, scoring=scores, return_train_score=True)
results_data = pd.DataFrame(results)
results_data
```

```python
print("PERFORMANCE MODEL ENSEMBLE DENGAN METODE VOTING")

accuracy = results_data.test_accuracy.mean() * 100
precision = results_data.test_precision.mean() * 100
recall = results_data.test_recall.mean() * 100

print("Accuracy : %0.2f" % accuracy, "%")
print("Precision : %0.2f " % precision, "%")
print("Recall : %0.2f " % recall, "%")
```
